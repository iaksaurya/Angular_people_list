###mean, median, standard deviation and gaussian distribution of int and float type columns

```python
from scipy.stats import norm

for col in affordable_housing.columns.values:#reach at each columns
  if affordable_housing[col].dtype != 'object':#check numerical type
      col_mean=affordable_housing[col].mean()#find mean of column
      col_median=affordable_housing[col].median()#find median of column
      satDev=affordable_housing[col].std()# find standard deviation
      # print( "mean,median and standard deviation corresponding to  year  is ",col_mean )
      print('mean , median and Standard Deviation  of year ' +str(col) + ' is ' +str(col_mean) + ' , ' +str(col_median) +" and " +str(satDev))

      x_values = np.linspace(col_mean - 4*satDev, col_mean + 4*satDev, 1000)# Generate values (x-axis)
      pdf_values = norm.pdf(x_values, col_mean, satDev)# Calculate the PDF using the Gaussian distribution
      plt.plot(x_values, pdf_values, label=f'Gaussian PDF\nMean={col_mean}, Std Dev={satDev}', color='blue')# Plot the Gaussian distribution
      # Add labels and title
      plt.xlabel('X values')
      plt.ylabel('Probability Density')
      plt.title('Gaussian (Normal) Distribution')
      # Show legend
      plt.legend()
      # Show the plot
      plt.grid(True)
      plt.show()
```

```python



# Convert 'ZIP' column in both DataFrames to string and ensure 5-digit format
affordable_housing['zip'] = affordable_housing['zip'].astype(str).str.zfill(5)
health_inspection['zip'] = health_inspection['zip'].astype(str).str.zfill(5)

# Define a function to use sqldf
pysqldf = lambda q: sqldf(q, globals())

#SQL query to join the dataframes
query = """
  SELECT
    pv.*,
    ppv.*
  FROM
    affordable_housing AS pv
  LEFT JOIN
    health_inspection AS ppv
  ON
    pv.zip = ppv.zip;
"""



# Execute the query
joined_data = pysqldf(query)

# Display the joined data
joined_data.columns


```

# E. Formulate and test the hypothesis:
 i. Formulate a reasonable alternative hypothesis
 ii. Formulate a null hypothesis
 iii. Select an appropriate test and significance level
 iv. Perform the test and decide if the null hypothesis should be rejected and alternative hypothesis should be accepted


```python
from sklearn.linear_model import LinearRegression#import library for LinearRegression
joined_data.fillna(0, inplace=True)

X = joined_data[['PROJECT_TOTAL_UNITS']]  # Predictor variable
y = joined_data['score']  # Predicted variable

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Display the coefficients and p-value
print(f"Coefficient: {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")

# Predict scores
y_pred = model.predict(X)
print(f"Predicted scores: {y_pred}")

# Calculate R-squared value
r_squared = model.score(X, y)
print(f"R-squared: {r_squared}")

# Since sklearn does not provide p-values directly, consider using statsmodels for p-values
import statsmodels.api as sm

X_with_const = sm.add_constant(X)  # Adds a constant term to the predictor
model = sm.OLS(y, X_with_const).fit()
print(model.summary())

```
